# 1. Feature Classification: Categorical vs Numerical

### Which features are categorical ?
- Categorical feature: nominal, ordinal, ratio, interval


### Which features are numerical ?
- Numerical: discrete - Dữ liệu rời rác, continuous - liên tục, timeseries - thời gian

### Phân loại features
1. Categorical: `Survived`, `Sex`, `Embarked`, `Pclass` (oridinal), `SibSp`, `Parch`
2. Numeric: `Age` (continuous), `Fare`(discrete)
3. Mix type of data: `Ticket`, `Cabin`
4. Contain Error/Typo
5. Blank or Null: `Cabin` > `Age` > `Embarked`
6. Various Data Type: int, String, Float

### Casting type cho các kiểu dữ liệu về Categorical

- [Survived]

```python
train_df['Survived'] = train_df['Survived'].astype("category")
```

```python
train_df.Survived.dtype
```

```python
train_df.info()
```

- Categorical: Survived, Sex, Embarked, Pclass (oridinal), SibSp, Parch

```python
categorical_features = ['Pclass','Embarked','Sex','SibSp','Parch','Survived']
```

```python
def convert_categorical(df, features):
    for feature in features:
        df[feature] = df[feature].astype('category')
        
convert_categorical(train_df, categorical_features)
convert_categorical(test_df, categorical_features)
```
